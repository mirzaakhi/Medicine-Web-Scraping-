```python
import requests, io, xlsxwriter
from bs4 import BeautifulSoup
from xml.sax.saxutils import escape
escape("<div>")
source = ["https://medex.com.bd/companies/2/aci-limited", "https://medex.com.bd/companies/2/aci-limited?page=2", "https://medex.com.bd/companies/2/aci-limited?page=3"]
for i in range(0,3):
    URL = source[i]
    headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'}
    html_page = requests.get(URL).text
    filename = "sheet"+str(i)+".csv"
    soup = BeautifulSoup (html_page, 'html.parser')
    #medicine = soup.find_all("div", { "class" : "col-xs-12 data-row-top" })
    medicine = soup.find_all('div', class_ = 'col-xs-12 data-row-top')
    
    open(filename, "w").close()
    for item in medicine:
        with io.open(filename, "a",encoding="utf-8") as worksheet:
                worksheet.write(str(item.get_text()))
        worksheet.close()
```
